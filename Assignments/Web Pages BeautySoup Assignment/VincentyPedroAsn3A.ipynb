```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
url = "https://home.treasury.gov/resource-center/data-chart-center/interest-rates/TextView?type=daily_treasury_yield_curve&field_tdr_date_value=2023"
```

```python
dfs = pd.read_html(url)
```

```python
len(dfs)
```

```python
df = dfs[0]
df.set_index("Date",inplace = True)
```

```python
df = df.iloc[:, 9:]
df
```

```python
y = df.iloc[-1].tolist()
```

```python
x = (1/12,2/12,3/12,4/12,6/12,1,2,3,5,7,10,20,30)
```

```python
plt.scatter(x,y,color = "maroon")
plt.title("Treasury Yield Curve")
plt.xlabel("Years")
plt.ylabel("Rates")
plt.show()
```

```python
df[["2 Yr","5 Yr","10 Yr"]].plot()
plt.xticks(rotation=60)
plt.ylabel("Rates")
plt.show()
```
