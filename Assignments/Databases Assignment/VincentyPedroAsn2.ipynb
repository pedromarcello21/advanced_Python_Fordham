```python
import pandas as pd
import sqlite3 as db
```

## Part A

```python
# Establish connection
connection = db.connect("dj.db")

# Define cursor to submit requests and hold results
cursor = connection.cursor()

#Run SQL Statement
cursor.execute('SELECT id, ticker, name from tickers')

# Get result of SQL statement request
rows = cursor.fetchall()

# Implement pandas
df = pd.DataFrame(rows)

#Establish column names
df.columns =['id','ticker','name']

#Set index
df.set_index('ticker', inplace = True)

df
```

## Part C

## Part D

```python
#Write Query
query = """
SELECT a.ticker,a.name,b.opening,b.closing,PRINTF('%1.2f%%',(b.closing-b.opening)/b.opening*100) AS change 
FROM tickers a left join daily_measures b on a.id = b.ticker_id WHERE ticker = 'AAPL'
"""

#Reveal Query Result
pd.read_sql_query(query,connection)
```
