```python
import yahoo_fin.options as ops
import yahoo_fin.stock_info as si
import opstrat as op
import datetime
import pandas as pd
```

```python
today = datetime.date.today()
formatted_date = today.strftime('%m/%d/%Y')
formatted_date
```

```python
expiration_dates = ops.get_expiration_dates("AAPL")
#expiration_dates
```

```python
AAPLCalls = ops.get_calls("AAPL",expiration_dates[-3])
AAPLCalls.set_index("Contract Name",inplace = True)
#AAPLCalls
```

```python
AAPLPuts = ops.get_puts("AAPL",expiration_dates[-3])
AAPLPuts.set_index("Contract Name",inplace = True)
AAPLPuts
```

### Single Plot

```python
AAPLCall = AAPLCalls.loc['AAPL250117C00160000']
AAPLCall
```

```python
AAPL_daily = si.get_data("AAPL", start_date="01/01/2023", end_date=today, index_as_date = True, interval="1d")
AAPL_today = AAPL_daily.iloc[-1]
AAPL_today
```

```python
AAPLSpot = AAPL_today[3]
AAPLStrike = AAPLCall[1]
```

```python
op.single_plotter(spot=AAPLSpot, strike=AAPLStrike, op_type='c', tr_type='s', op_pr=AAPLCall[2])
```

### Multi Plot (Portfolio)

```python
AAPLPut = AAPLPuts.loc['AAPL250117P00140000']
AAPLPut
```

```python
AAPLStrike1 = AAPLPut[1]
AAPLStrike2 = AAPLCall[1]
```

```python
op1={'op_type': 'p', 'strike': AAPLStrike1, 'tr_type': 's', 'op_pr': AAPLCall[2]}
op2={'op_type': 'c', 'strike': AAPLStrike2, 'tr_type': 'b', 'op_pr': AAPLPut[2]}

op_list=[op1, op2]
op.multi_plotter(spot=AAPLSpot, op_list=op_list)
```

```python
from optionprice import Option

ploss = {'Equity Market Price at Expiry': [AAPLStrike2-10,AAPLStrike2,AAPLStrike2+10,AAPLStrike2+20,
                                           AAPLStrike2+30,AAPLStrike2+40]}
noc = int(AAPLCall[-3]) #number of contracts bought/sold
sperc = 100*noc #100 shares per contract
df = pd.DataFrame(ploss, columns = ['Equity Market Price at Expiry','Profit/Loss'])
payoffs = []
for i in df['Equity Market Price at Expiry']:

# Calculate Option Payoffs
        payoff = ((i - AAPLStrike2) * sperc) - (AAPLCall[2]*sperc)
        payoffs.append(payoff)

# Create DataFrame
ploss['Profit/Loss']=payoffs
final=pd.DataFrame(ploss,columns=['Equity Market Price at Expiry','Profit/Loss']).set_index('Equity Market Price at Expiry')

final
```
