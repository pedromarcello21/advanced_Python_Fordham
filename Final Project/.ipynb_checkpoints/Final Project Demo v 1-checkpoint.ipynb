```python
import yahoo_fin.options as ops
import yahoo_fin.stock_info as si
import opstrat as op
import datetime
```

```python
today = datetime.date.today()
formatted_date = today.strftime('%m/%d/%Y')
formatted_date
```

```python
expiration_dates = ops.get_expiration_dates("DIS")
#expiration_dates
```

```python
disneyCalls = ops.get_calls("DIS",expiration_dates[-3])
disneyCalls.set_index("Contract Name",inplace = True)
#disneyCalls
```

```python
disneyPuts = ops.get_puts("DIS",expiration_dates[-3])
disneyPuts.set_index("Contract Name",inplace = True)
#disneyPuts
```

### Single Plot

```python
disneyCall = disneyCalls.loc['DIS240621C00100000']
disneyCall
```

```python
dis_daily = si.get_data("dis", start_date="01/01/2023", end_date=today, index_as_date = True, interval="1d")
dis_today = dis_daily.iloc[-1]
dis_today
```

```python
disSpot = dis_today[3]
disStrike = disneyCall[1]
```

```python
op.single_plotter(spot=disSpot, strike=disStrike, op_type='c', tr_type='s', op_pr=disneyCall[2])
```

### Multi Plot (Portfolio)

```python
disneyPut = disneyPuts.loc['DIS240621P00100000']
```

```python
disSpot = dis_today[3]
disStrike1 = disneyPut[1]
disStrike2 = disneyCall[1]
```

```python
op1={'op_type': 'c', 'strike': disStrike1, 'tr_type': 's', 'op_pr': disneyCall[2]}
op2={'op_type': 'p', 'strike': disStrike2, 'tr_type': 's', 'op_pr': disneyPut[2]}

op_list=[op1, op2]
op.multi_plotter(spot=disSpot, op_list=op_list)
```
