```python
# Import relevant packages
import requests
import os
from dotenv import load_dotenv
import pandas as pd
```

```python
# get a previously issued AlphaVantage API key from an .env file
# *Pedro's key name is 'alphaVantageKey'.  Key name varies by user

env_file = 'env_vars.env'
load_dotenv(env_file)
myAPIKey = os.getenv('alphaVantageKey')
#print(myAPIKey)
```

```python
# Establish stock
symbol = 'RKLB'

# define the url where the web service is
API_URL = "https://www.alphavantage.co/query"

# define the details for the Alpha Vantage request (key/value pairs in a dictionary)
data = {
        "function": "TIME_SERIES_INTRADAY",
        "symbol": symbol,
        "interval": "5min",
        "datatype": "json",
        "apikey": myAPIKey
       }
```

```python
# Issue request to web service
response = requests.get(API_URL, data)
print(type(response))
```

```python
# Convert requested data from web service to a python dictionary
data = response.json()
```

```python
# Create data frame from dictionary
time_series = (data['Time Series (5min)'])
df = pd.DataFrame.from_dict(time_series)
dfRKLB = df.transpose()
dfRKLB.index.name = 'Date_Time'
dfRKLB
```

```python
# Establish Stock
symbol = 'ASTS'

# define the url where the web service is
API_URL = "https://www.alphavantage.co/query"

# define the details for the Alpha Vantage request (key/value pairs in a dictionary)
data = {
        "function": "TIME_SERIES_INTRADAY",
        "symbol": symbol,
        "interval": "5min",
        "datatype": "json",
        "apikey": myAPIKey
       }
```

```python
# Issue request to web service
response = requests.get(API_URL, data)
```

```python
# Convert requested data from web service to a python dictionary
data = response.json()
```

```python
# Create data frame from dictionary
time_series = (data['Time Series (5min)'])
df = pd.DataFrame.from_dict(time_series)
dfASTS = df.transpose()
dfASTS.index.name = 'Date_Time'
```

```python
# Consolidate data from both stocks onto one dataframe
dfSpaceStocks = pd.merge(dfRKLB,dfASTS,on = 'Date_Time',how = 'inner')
dfSpaceStocks
```

```python
# Column Type Check
dfSpaceStocks.info()
```

```python
# Set index column from string to datetime type
dfSpaceStocks.index = dfSpaceStocks.index.astype("datetime64[ns]")
```

```python
# Rename Columns
cols = ['Open_RKLB','High_RKLB','Low_RKLB','Close_RKLB','Volume_RKLB','Open_ASTS','High_ASTS','Low_ASTS','Close_ASTS','Volume_ASTS']
dfSpaceStocks.columns = cols
```

```python
# Convert columns to numeric type
dfSpaceStocks[cols] = dfSpaceStocks[cols].apply(pd.to_numeric)
dfSpaceStocks = dfSpaceStocks.round(2)
```

```python
# Plot the closing prices of both stocks
dfTest = dfSpaceStocks[['Close_RKLB','Close_ASTS']]
%matplotlib inline
dfTest.plot.line(subplots=False,xlabel = "Date_Time (MM-YY HH)")
```

```python
# Correlation between RocketLab USA and ASTS Mobile
print(dfTest['Close_RKLB'].corr(dfTest['Close_ASTS']).round(5))
```
