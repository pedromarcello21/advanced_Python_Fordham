```python
df['Country'] = df['Country'].apply(lambda x: x.rstrip('[i],[ii],[iii],[iv]'))
df['Country']
```

```python
#Function to drop footnotes and commas in df
def dropFootnoteAndCommas(s):
    result = ''
    for c in s:
        if c == '[': 
            break
        if c!= ',':
            result +=c
    return result
```

```python
#Check
dropFootnoteAndCommas('1,234[a]')
```

```python
#for i in df['Country']:
#    for j in i:
#        print(j)
```

```python
df['Country'] = df['Country'].apply(dropFootnoteAndCommas)
```

```python
df['Country']
```

```python
df['Sales2021'] = df['Sales2021'].apply(dropFootnoteAndCommas)
```

```python
df['Sales2020'] = df['Sales2020'].apply(dropFootnoteAndCommas)
```
