```python
import pandas as pd
df = pd.read_csv("DQ-Example-Strings.txt", sep='|', index_col='id')
df
```

```python
df[['city', 'state']] = df.address.str.split(',', expand=True)
df
```

```python
df['state'] = df['state'].str.strip()
for index, row in df.iterrows():
   print(row['state'])
```

```python
dfstates = pd.read_csv('states.txt')
stateList = dfstates.state.tolist()
stateList
```

```python
df['state_ok'] = df.state.apply(lambda x: x in stateList)
df
```

```python
df['state_clean'] = \
   df.state.apply(lambda x: x if x in stateList else None)
df
```
