*** requires datafiles: contacts.csv, phones.csv

```python
import pandas as pd
dfContacts = pd.read_csv('contacts.csv', index_col='id')
dfContacts
```

```python
dfPhones = pd.read_csv('phones.csv', index_col='id')
dfPhones
```

```python
dfJoin = dfContacts.join(dfPhones)
dfJoin
```

```python
dfJoinOuter = dfContacts.join(dfPhones, how='outer')
dfJoinOuter
```

```python
dfJoinOuter[dfJoinOuter['phone'].isna()]
```

```python
dfJoinOuter[dfJoinOuter['name'].isna()]
```
