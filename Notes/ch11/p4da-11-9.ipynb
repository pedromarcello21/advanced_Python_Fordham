```python
import pandas as pd
df = pd.read_csv('popgdp.csv')
df
```

```python
df['gdppercap'] = round(df.gdp / df.population * 1000000)
df
```

```python
gdppercap = \
   df.pivot(index='year', columns='country', values='gdppercap')
gdppercap
```

```python
gdppercap.transpose()
```
