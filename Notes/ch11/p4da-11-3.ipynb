```python
import pandas as pd
df = pd.read_csv("DQ-Example-Dates.csv", index_col='id')
df
```

```python
df['date_clean'] = df['date'].str.replace(r'[^0-9/-]','', regex=True)
df[['date', 'date_clean']]
```

```python
df['date_clean'] = pd.to_datetime(df['date_clean'], errors='coerce')
df
```

```python
dfnona = df.dropna(subset=['date_clean'])
dfnona
```
