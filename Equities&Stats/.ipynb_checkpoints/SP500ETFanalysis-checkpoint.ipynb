#### A) Read the 12 datafiles

```python
import pandas as pd
```

```python
# the tickers of the ETFs
ETF_list = ['XLC', 'XLY', 'XLP', 'XLE', 'XLF', 'XLV', 'XLI', 'XLB', 'XLRE', 'XLK', 'XLU', 'SPY']
```

```python
# The dictionary for mapping the tickers in the previous cell to sector names
ETF_dict = {}
ETF_dict['XLC']  = 'Comm'
ETF_dict['XLY']  = 'ConDis'
ETF_dict['XLP']  = 'ConSta'
ETF_dict['XLE']  = 'Energy'
ETF_dict['XLF']  = 'Fin'
ETF_dict['XLV']  = 'Health'
ETF_dict['XLI']  = 'Indus'
ETF_dict['XLB']  = 'Mater'
ETF_dict['XLRE'] = 'RealE'
ETF_dict['XLK']  = 'Tech'
ETF_dict['XLU']  = 'Util'
ETF_dict
```

```python
# read the 1st ETF
ETF_raw = pd.read_csv('XLC.csv', parse_dates=['Date'])[['Date', 'Adj Close']].set_index('Date').rename(columns = {'Adj Close': 'XLC'})
```

```python
# read the rest of the ETFs by for loop
for ETF in ETF_list[1:]:
    tmp = pd.read_csv(ETF + '.csv', parse_dates=['Date'])[['Date', 'Adj Close']].set_index('Date').rename(columns = {'Adj Close': ETF})
    ETF_raw = pd.merge(ETF_raw, tmp, how = 'outer', left_index = True, right_index = True)
```

```python
# map ETFs to sectors and calculate percentage returns (%)
ETF_raw = ETF_raw.rename(columns = ETF_dict)
#pd.set_option('precision', 2)
pd.set_option('display.precision', 2)
ETF_ret = ETF_raw.pct_change() * 100
ETF_ret = ETF_ret.iloc[1:]
ETF_ret.head(10)
```

#### B) Get a simple statistic review of the returns: mean, stddev

```python
# Extract the mean & standard deviation of ETFs
ETF_stats = ETF_ret.describe()
ETF_stats_raw = ETF_stats[(ETF_stats.index == 'mean') | (ETF_stats.index == 'std')].T
ETF_stats_raw.sort_values(by = ['mean'], inplace = True)
ETF_stats_sorted = ETF_stats_raw
ETF_stats_sorted.T.round(2)
```

```python
import matplotlib.pyplot as plt
```

```python
# bar chart of the mean & stddev of ETFs
f, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 4))
ax1.barh(ETF_stats_sorted.index, ETF_stats_sorted['mean'].values)
ax1.set_xlabel('Average daily return(%)')
ax1.set_ylabel('Sectors')
ax2.barh(ETF_stats_sorted.index, ETF_stats_sorted['std'].values)
ax2.set_xlabel('Average daily standard deviation(%)')
ax2.set_ylabel('Sectors')
```

#### D) Review the 1yr, 3yr, 5yr returns of the sectors

```python
# Given the list of dates, and the number of years to look back,
# this function returns the start date and end date for calculating nominal returns.
# Note the returned start date is adjusted to be the closest earlier business day.
from dateutil.relativedelta import relativedelta

def get_start_date(date_list, nyears):   
            
    end_date= date_list[-1].date()
    timedelta = relativedelta(years=nyears)
    start_date = pd.Timestamp(end_date) - timedelta

    # check if the lookback goes off the end of the provided date series
    #print(start_date, date_list[0])
    if start_date < date_list[0]:
        raise Exception("Sorry, the start date is out of range.")

    # move back to a prior business day
    i = 0
    while start_date not in date_list: # TODO - make more efficient
        i += 1
        start_date = pd.Timestamp(end_date) - timedelta - pd.Timedelta(str(i)+' D')
            
    start_date = start_date.date()
    print("period:", start_date, '-', end_date)
    return start_date, end_date
```

```python
# test invalid input
try:
    get_start_date(ETF_raw.index, 10)
except Exception as err:
    print('There is an error:', err)
```

```python
# calculate nominal returns given data and desired time range (in years)
def calcNominalReturn(data, nyears):
    start_date, end_date =  get_start_date(data.index, nyears)
    return data[data.index == pd.Timestamp(end_date)].values/data[data.index == pd.Timestamp(start_date)].values - 1
```

```python
# calculate 1-year annualized returns
anl1yr = calcNominalReturn(ETF_raw, 1)
```

```python
# calculate 1-year annualized returns
anl1yr = calcNominalReturn(ETF_raw, 1)
```

```python
# calculate 3-year annualized returns
nominal3yr = calcNominalReturn(ETF_raw, 3)
anl3yr = (1+nominal3yr)**(1/3) -1
```

```python
# calculate 5-year annualized returns
nominal5yr = calcNominalReturn(ETF_raw, 5)
anl5yr = (1+nominal5yr)**(1/5) -1
```

```python
# integrate the returns to one dataframe
# note: .T transposes (since the original data has tickers across, and we are creating a result w/ tickers down)
anl_ret = pd.DataFrame(anl1yr.T, index = ETF_raw.columns, columns = ['1yr'])
anl_ret['3yr'] = anl3yr.T
anl_ret['5yr'] = anl5yr.T
anl_ret = anl_ret * 100
anl_ret.round(1)
```

```python
# plot heatmap of ETF 1/3/5yr annual returns
plt.figure(figsize = (6, 6))
rdgn = sns.diverging_palette(h_neg=13, h_pos=133, s=80, l=55, sep=3, as_cmap=True)
sns.heatmap(anl_ret, cmap=rdgn, annot = True, fmt = '.1f', center = 0, vmin = -32)
```

```python
# get subset of DataFrame on/after a given date
def dfAfter(df, startDate):
    result = df[startDate:]
    return result
```

```python
start_date, end_date = get_start_date(ETF_raw.index, 1)
dfAfter(ETF_raw, start_date).head()
```

#### E) Visualize the daily returns of all sectors

```python
# get data to prepare for n-year performance time series chart
nyears = 1
start_date, end_date = get_start_date(ETF_raw.index, nyears)
# get latest nyears of data
ETF_px_sub = dfAfter(ETF_raw, start_date)
# normalize to 100
ETF_px_sub100 = ETF_px_sub / ETF_px_sub.iloc[0] * 100
ETF_px_sub100.head()
```

```python
# calculate and chart the 1 year daily performance of ETFs, start=100
fig, ax = plt.subplots(figsize = (12, 6))
ETF_px_sub100.plot(ax = ax)
# TODO: highlight the SPY series, e.g., wide black
ax.legend(loc='upper left', ncol=2)
plt.title('ETF performances')
plt.ylabel('price')
plt.gca().spines['top'].set_color(None)
plt.gca().spines['right'].set_color(None)
```

#### F) Analyze the relationships between sectors and S&P500 w/ Betas

```python
ETF_ret.head(10)
```

```python
# Extract data for regressions between ETF setors and the S&P500
from scipy import stats
nyears = 1
start_date, end_date = get_start_date(ETF_raw.index, nyears)
ETF_ret_regress = dfAfter(ETF_ret, start_date)
ETF_ret_regress.head()
```

```python
# Perform regressions between ETF sectors and S&P500 (daily returns)
sector_list = []
intercept_list = []
slope_list = []
r_value_list = []
p_value_list = []
for i in ETF_ret_regress.columns[:-1]:
    slope, intercept, r_value, p_value, std_err = stats.linregress(ETF_ret_regress['SPY'].values, ETF_ret_regress[i].values)
    sector_list.append(i)
    intercept_list.append(intercept)
    slope_list.append(slope)
    r_value_list.append(r_value)
    p_value_list.append(p_value)
```

```python
# Summarize the regression results in a DataFrame
Reg_report={'sector':sector_list,'alpha':intercept_list,
            'beta':slope_list,'r_value':r_value_list,
            'p_value':p_value_list}
Reg_report_df = pd.DataFrame(data = Reg_report,
                             columns=['sector','alpha','beta','p_value', 'r_value'])
Reg_report_df['r2'] = Reg_report_df['r_value'] ** 2 * 100
Reg_report_df.drop('r_value', axis=1, inplace=True)
Reg_report_df.round({"alpha":2, "beta":2, "p_value":3, "r2": 1})
```
